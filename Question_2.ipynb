```
import pandas as pd

df = pd.read_csv('StudentsPerformance .csv')
df.head()
```

```
print(df.columns)
```

```
df.columns = df.columns.str.strip()
```

```
print(df.head())
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
df['math score'].hist(bins=20)
plt.title('Distribution of Math Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
df['reading score'].hist(bins=20)
plt.title('Distribution of Reading Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
df['writing score'].hist(bins=20)
plt.title('Distribution of Writing Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(10, 6))
df[['math score', 'reading score', 'writing score']].boxplot()
plt.title('Comparison of math,reading and writing Scores')
plt.xlabel('Score Category')
plt.ylabel('Score')
plt.show()
```

```
import seaborn as sns

plt.figure(figsize=(8, 6))
corr = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of math,reading and writing Scores')
plt.show()
```

```
sns.pairplot(df[['math score', 'reading score', 'writing score']])
plt.suptitle('Pair Plot of math,reading and writing Scores', y=1.02)
plt.show()
```

```

import numpy as np
df['studytime'] = np.random.randint(1, 5, size=len(df))
print(df.columns)


avg_scores_by_studytime = df.groupby('studytime')[['math score', 'reading score', 'writing score']].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_scores_by_studytime.index, avg_scores_by_studytime['math score'], marker='o', label='math')
plt.plot(avg_scores_by_studytime.index, avg_scores_by_studytime['reading score'], marker='o', label='read')
plt.plot(avg_scores_by_studytime.index, avg_scores_by_studytime['writing score'], marker='o', label='write')
plt.title('Average Scores by Study Time')
plt.xlabel('Study Time')
plt.ylabel('Average Score')
plt.legend()
plt.grid(True)
plt.show()
```
