Importing the libraries

```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# Define the path to the dataset
raw_file_path = "../dataraw/Assignment1.csv"
clean_file_path = "../dataclean/Cleaned_Assignment1.csv"
results_file_path = "../results/test_results.txt"
```

```python
# Read the raw data
raw_data = pd.read_csv(raw_file_path)
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(raw_data.head())

# Display basic information about the dataset
print("\nBasic information about the dataset:")
print(raw_data.info())

# Display summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(raw_data.describe())
```

```python


# Clean the data by removing rows with NA values
clean_data = raw_data.dropna()

# Check for duplicate rows and remove them
print("\nNumber of duplicate rows:")
print(clean_data.duplicated().sum())
clean_data = clean_data.drop_duplicates()

# Display cleaned dataset information
print("\nCleaned dataset information:")
print(clean_data.info())

```

```python
# Save the cleaned data to disk
clean_data.to_csv(clean_file_path, index=False)

print(f"\nCleaned dataset saved to {clean_file_path}")
```

```python
clean_data = pd.read_csv(clean_file_path)

# Perform a t-test to compare 'Grip Strength (kg)' between 'Frailty' groups
group_N = clean_data[clean_data['Frailty'] == 'N']['Grip Strength (kg)']
group_Y = clean_data[clean_data['Frailty'] == 'Y']['Grip Strength (kg)']

# Ensure groups have sufficient data
if len(group_N) > 1 and len(group_Y) > 1:
    t_test_result = stats.ttest_ind(group_N, group_Y)
else:
    raise ValueError("Not enough data in one of the groups for a t-test.")

# Write the test result to a plain text file
with open(results_file_path, 'w') as f:
    f.write(f"T-test result for Grip Strength between Frailty groups (N and Y):\n{t_test_result}")

print(f"T-test results saved to {results_file_path}")
```
